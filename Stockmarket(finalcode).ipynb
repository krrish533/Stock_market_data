***FINAL CODE FOR RETRIEVING 3 YEAR'S(2020,2021,2022) DATA OF (BANKNIFTY,NIFTY,RELIANCE)***

```python
import os
import zipfile
import pandas as pd

file_path = '/content/drive/MyDrive/NF BNF Data/oneminutedata'
years_input = input("Enter the years separated by commas (e.g., 2020,2021,2022): ")
years = years_input.split(',')
#years = [2020,2021,2022]
#company_name = 'NIFTY'
company_name = input("Enter the company name in Upper Case (e.g., BANKNIFTY,NIFTY,RELIANCE) : ")

dataframes = []

for year in years:
    year_path = os.path.join(file_path, str(year))
    if os.path.isdir(year_path):
        for month_folder in ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']:
            month_path = os.path.join(year_path, month_folder)
            if os.path.isdir(month_path):
                month_zip_path = os.path.join(month_path, f"IntradayData_{month_folder}{year}.zip")
                if os.path.isfile(month_zip_path):
                    with zipfile.ZipFile(month_zip_path, 'r') as zip_ref:
                        for file_name in zip_ref.namelist():
                            # Check if the file is a .txt file and contains the user-defined company name
                            if file_name.endswith('.txt') and company_name in file_name.upper():
                                # Check if the file name strictly matches the user-defined company name
                                file_name_clean = file_name.split('/')[-1].split('.')[0]
                                if file_name_clean == company_name:
                                    with zip_ref.open(file_name) as file:
                                        try:
                                            df = pd.read_csv(file, delimiter=',', header=None)
                                            df.columns = ['CompanyName', 'Date', 'Time', 'Open', 'High', 'Low', 'Close', 'SharesTraded', 'Turnover']
                                            dataframes.append(df)
                                        except Exception as e:
                                            print(f"Error reading file {file_name}: {e}")

if dataframes:
    final_dataframe = pd.concat(dataframes, ignore_index=True)
    print("Displaying DataFrame:")
    print(final_dataframe)

    #csv_file_path = f'/content/drive/MyDrive/extract_dir/{company_name}_data_{years_input}_ALL_MONTHS.csv'
    #final_dataframe.to_csv(csv_file_path, index=False)
    #print(f"Data saved to {csv_file_path}")
#else:
    #print(f"No data found for {company_name} in the specified year(s).")
```
